```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
#sample document
doc = "The quick brown foxes are jumping over the lazy dogs."
```

```python
#Tokenization
tokens = word_tokenize(doc.lower())
print(tokens)
```

```python
#stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [w for w in tokens if w.isalnum() and w not in stop_words]
print(filtered_tokens)
```

```python
#pos tags
pos_tags = nltk.pos_tag(filtered_tokens)
print(pos_tags)
```

```python
#stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered_tokens]
print(stemmed)
```

```python
#Lemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_tokens]
```

```python
print(f"Original: {doc}")
print(f"Lemmatized: {lemmatized}")
```

```python
corpus = [
    "The quick brown foxes are jumping over the lazy dogs",
    "Data science is amazing and data is power",
    "The lazy dogs are sleeping"
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)
```

```python
feature_names = vectorizer.get_feature_names_out()
first_doc_scores = tfidf_matrix.toarray()[0]
```

```python
tfidf_results = dict(zip(feature_names, first_doc_scores))
```

```python
print(tfidf_results)
```
